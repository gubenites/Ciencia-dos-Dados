```python
#Se usar o arquivo descompactado
dados=pd.read_csv('DOM2015.csv',sep=',')

#Se usar o arquivo compactado como está no dropbox púbico
dados2=pd.read_csv('DOM2014.csv', sep=',')
```

```python
pd.to_numeric(dados.UF)
dados.UF[(dados.UF >=10) & (dados.UF <=19)]= [10]
dados.UF[(dados.UF >=20) & (dados.UF <=29)]= [20]
dados.UF[(dados.UF >=30) & (dados.UF <=39)]= [30]
dados.UF[(dados.UF >=40) & (dados.UF <=49)]= [40]
dados.UF[(dados.UF >=50) & (dados.UF <=59)]= [50]

dados.UF=dados.UF.astype("category")
dados.UF.cat.categories=("Norte","Nordeste","Suldeste","Sul","Centro-Oeste")

dados.V02426=dados.V02426.astype("category")
dados.V02426.cat.categories=("Sim","Não")

dados.V02424=dados.V02424.astype("category")
dados.V02424.cat.categories=("Sim","Não")

pd.to_numeric(dados2.UF)
dados2.UF[(dados2.UF >=10) & (dados2.UF <=19)]= [10]
dados2.UF[(dados2.UF >=20) & (dados2.UF <=29)]= [20]
dados2.UF[(dados2.UF >=30) & (dados2.UF <=39)]= [30]
dados2.UF[(dados2.UF >=40) & (dados2.UF <=49)]= [40]
dados2.UF[(dados2.UF >=50) & (dados2.UF <=59)]= [50]

dados2.UF=dados2.UF.astype("category")
dados2.UF.cat.categories=("Norte","Nordeste","Suldeste","Sul","Centro-Oeste")

dados2.V02426=dados2.V02426.astype("category")
dados2.V02426.cat.categories=("Sim","Não")

dados2.V02424=dados2.V02424.astype("category")
dados2.V02424.cat.categories=("Sim","Não")

ct3=pd.crosstab([dados2.UF],[dados2.V02426],normalize=True)
soma3=ct3.sum().sum()
ct3*100/soma3
ct3 = (ct3/ct3.sum().sum()*100).round(decimals=5)
ct3.plot(kind='bar',color=('blue','red'),)
plt.ylabel('Frequencia relativa em %')
plt.title('Acesso a internet por banda larga em 2014')

ct1=pd.crosstab([dados.UF],[dados.V02426],normalize=True)
soma2=ct1.sum().sum()
ct1*100/soma2
ct1 = (ct1/ct1.sum().sum()*100).round(decimals=5)
ct1.plot(kind='bar',color=('blue','red'))
plt.ylabel('Frequencia relativa em %')
plt.title('Acesso a internet por banda larga em 2015')

ct=pd.crosstab(dados2.UF,dados2.V02424, normalize=True)
soma=ct.sum().sum()
ct*100/soma
ct = (ct/ct.sum().sum()*100).round(decimals=5)
ct.plot(kind='bar',color=('blue','red'))
plt.ylabel('Frequencia relativa em %')
plt.title('acesso a internet por 3G/4G em 2014')

ct=pd.crosstab([dados.UF],[dados.V02424], normalize=True)
soma=ct.sum().sum()
ct*100/soma
ct = (ct/ct.sum().sum()*100).round(decimals=5)
ct.plot(kind='bar',color=('blue','red'))
plt.ylabel('Frequencia relativa em %')
plt.title('acesso a internet por 3G/4G em 2015')


```
